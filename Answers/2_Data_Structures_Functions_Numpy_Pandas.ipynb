1.	Create a list of 5 integers. Perform and print the result of the following operations: append, extend, insert, remove, pop, clear, index, count, sort, and reverse.

```python
sample = [2,4,5,8,6]
sample.append(10)
print("List after appending : ", sample)
sample.extend([11,12])
print("List after extending : ", sample)
sample.insert(1,3)
print("List after inserting : ", sample)
sample.remove(2)
print("List after removing : ", sample)
sample.pop(len(sample)-1)
print("List after popping : ", sample)
print("Index of 11 : ", sample.index(11))
print("Count of 4 : ", sample.count(4))
sample.sort()
print("List after sorting : ", sample)
sample.reverse()
print("List after reversing : ", sample)
sample.clear()
print("List after clearing : ", sample)
```

6.	Given a 'marks' variable, print the grade:
- 90 and above: 'A'
- 75–89: 'B'
- 50–74: 'C'
- Below 50: 'Fail'

```python
def grade(marks):
    if marks >=90:
        return "A"
    elif (marks>= 75) and (marks<=89):
        return 'B'
    elif (marks>= 50) and (marks<=74):
        return 'c'
    return 'Fail'
print(grade(49))
```

7.	Given a number, check if it's positive, and if it is also even. If not positive, print if it's zero or negative.

```python
def check(n):
    if n>=0 and n%2==0:
        print(n,"is positive and even")
    elif n>=0 and n%2 !=0:
        print(n,"is positive and odd")
    elif n<0 and n%2 ==0:
        print(n,"is Negative and even")
    elif n<0 and n%2 !=0:
        print(n,"is Negative and odd")
    elif n==0:
        print(n,"is zero")
check(-4)
```
