Exercise 1: Prime Numbers
Write a Python program that checks whether a given number is prime or not. A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.


```python
def prime(x):
    if x <= 1:
        return False  # 0 and 1 are not prime
    if x == 2:
        return True   # 2 is prime
    if x % 2 == 0:
        return False  # Even numbers > 2 are not prime
    for i in range(3, int(x**0.5) + 1, 2):  # Check odd divisors up to sqrt(x)
        if x % i == 0:
            return False
    return True

print(f"prime(17) = {prime(17)}")  # True
print(f"prime(18) = {prime(18)}")  # False
```

Exercise 2: Product of Random Numbers
Develop a Python program that generates two random numbers and asks the user to enter the product of these numbers. The program should then check if the user's answer is correct and display an appropriate message.


```python
import random

def product_game():
    num1 = random.randint(1, 10)
    num2 = random.randint(10, 20)
    expected_product = num1 * num2
    
    try:
        user_answer = int(input(f"Enter the product of {num1} x {num2}: "))
        if user_answer == expected_product:
            print(f"Correct! {num1} x {num2} = {expected_product}")
        else:
            print(f"Wrong! The correct answer is {expected_product}")
    except ValueError:
        print("Invalid input. Please enter a valid integer.")


product_game()
```

✅ Problem Statement (Question): Hospital Billing System
A hospital charges its patients based on the following:
1. Room charges per day: ₹2,000
2. Doctor consultation fee (flat): ₹1,500
3. Lab test charges: Based on the number of tests taken. Each test costs ₹300
4. Medicine charges: Total cost of medicines provided
5. Discount: If the total bill (before discount) exceeds ₹10,000, a 10% discount is applied
##### Write a Python program to calculate the final bill for a patient who:
* Stayed for 4 days
* Had 3 lab tests
* Had medicine charges of ₹2,400


```python
def hospital_bill(days,lab_tests,medicine_charges):
    total_amount = 0
    total_amount = days *2000 + 1500 + 300*lab_tests + medicine_charges
    if total_amount >10000:
        total_amount = total_amount-( total_amount * 0.10)
    return total_amount
print("Final Bill : ",hospital_bill(4,3,2400))
```
