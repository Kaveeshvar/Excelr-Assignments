IMPUTATIONN STRATEGY :

```python
from sklearn.impute import SimpleImputer
```

```python
num_cols=adult.select_dtypes(include=["int64","float64"]).columns
cat_cols=adult.select_dtypes(include=["object"]).columns
```

```python
num_imputer=SimpleImputer(strategy="median")
cat_imputer=SimpleImputer(strategy="most_frequent")
```

```python
adult[num_cols]=num_imputer.fit_transform(adult[num_cols])
adult[cat_cols]=cat_imputer.fit_transform(adult[cat_cols])
```

Feature Scaling :

Standard Scaling :

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
adult_standard_scaled = adult.copy()
adult_standard_scaled[num_cols] = std_scaler.fit_transform(adult[num_cols])
```

Min-Max Scaling :

```python
from sklearn.preprocessing import MinMaxScaler

mm_scaler = MinMaxScaler()
adult_minmax_scaled = adult.copy()
adult_minmax_scaled[num_cols] = mm_scaler.fit_transform(adult[num_cols])
```

```python
# Standard Scaling :
#Preferred When	: During Linear Regression, Logistic Regression, SVM, PCA
#Reason : Centers data (mean=0, std=1), assumes Gaussian distribution


#Min-Max Scaling :
#Preferred When	: Neural Networks, KNN, Distance-based models
#Reason : Keeps values in fixed range [0,1], preserves distribution
```

Encoding Techniques :

One-Hot Encoding :

```python
adult_encoding = pd.get_dummies(adult, columns=["workclass","education","marital_status","occupation","relationship","race","sex","native_country"], drop_first=True)
adult_encoding
```

Label Encoding :

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
adult["education_encoded"] = le.fit_transform(adult["education"])
adult[["education", "education_encoded"]].head()
```

```python
cat_cols
```

```python
#Pros & Cons of Encoding Techniques

#One-Hot Encoding :

#Pros :

#No ordinal relationship assumed
#Works well for nominal data

#Cons :

#Increases dimensionality
#Not scalable for high-cardinality features

#Label Encoding :

#Pros :

#Memory efficient
#Fast and simple

#Cons :

#Introduces false ordinal relationship
#Can mislead distance-based models
```

Feature Engineering :

Feature 1: Capital Gain Indicator :

```python
adult["has_capital_gain"] = (adult["capital_gain"] > 0).astype(int)
```

```python
#1 : individual has a capital gain
#0 : no capital gain
#Capital gains are strongly associated with higher income levels.
```

Rationale :

```python
#Capital gains are strongly associated with higher income levels. Creating a binary indicator helps the model:
#Capture the presence or absence of capital gains
#Reduce sensitivity to extreme capital gain values
#Improve interpretability and predictive power
```

Feature 2: Work Intensity :

```python
adult["work_intensity"] = adult["hours_per_week"] * adult["education_num"]
```

```python
#hours-per-week : working effort
#education-num : education level
#Income is influenced not just by working hours or education individually, but by their combined effect.
```

Rationale :

```python
#Income is influenced not just by working hours or education individually, but by their combined effect.
#The work_intensity feature represents how education and work commitment together impact earning potential.
```

Transformation of Skewed Features :

```python
adult["capital_gain"].skew()       # check skew
#The capital-gain feature is highly right-skewed, meaning most values are small while a few extremely large values dominate the distribution
```

Log Transformation :

```python
adult["capital_gain_log"] = np.log1p(adult["capital_gain"])
```

```python
#Observation:

#The capital-gain feature is highly right-skewed, meaning most values are small while a few extremely large values dominate the distribution.
```

Justification :

```python
#Log transformation is applied to handle skewness effectively

#Reduces the impact of extreme outliers

#Makes the distribution more symmetric

#Improves performance of linear and distance-based models

#log1p() is used to safely handle zero values
```
