Task 2: Dimensionality Reduction with PCA :

Standardization :

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_Scaled=scaler.fit_transform(wine)
```

PCA & Explained Variance :

```python
from sklearn.decomposition import PCA
pca=PCA()
X_pca=pca.fit_transform(X_Scaled)
```

```python
wine.var().nlargest(2)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Scree Plot â€“ Wine Dataset")
plt.grid()
plt.show()
```

```python
#95% variance is captured using 5 principal components
```

PCA Transformation :

```python
pca_2 = PCA(n_components=2)
X_pca_2 = pca_2.fit_transform(X_Scaled)
```

Task 3: Clustering with Original Data :

K-Means Clustering :

```python
from sklearn.cluster import KMeans
kmeans_original = KMeans(n_clusters=3, random_state=42)
```

```python
labels_original = kmeans_original.fit_predict(X_Scaled)
```

Visualisation :

```python
plt.figure(figsize=(7, 5))
plt.scatter(X_Scaled[:, 0], X_Scaled[:, 1], c=labels_original, cmap='coolwarm')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Clustering")
plt.show()
```

Evaluation Metrics :

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score
sil=silhouette_score(X_Scaled,labels_original)
dav=davies_bouldin_score(X_Scaled,labels_original)

sil,dav
```

Task 4: Clustering with PCA Data :

```python
kmeans_pca = KMeans(n_clusters=3, random_state=42)
```

```python
labels_pca = kmeans_pca.fit_predict(X_pca_2)
```

Visualisation :

```python
plt.figure(figsize=(7, 5))
plt.scatter(X_pca_2[:, 0], X_pca_2[:, 1], c=labels_pca, cmap='coolwarm')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Clustering")
plt.show()
```

Evaluation Metrics :

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score
sil_pca=silhouette_score(X_pca_2,labels_pca)
dav_pca=davies_bouldin_score(X_pca_2,labels_pca)

sil_pca,dav_pca
```
